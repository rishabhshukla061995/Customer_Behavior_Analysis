### Checking the missing values

```python
df.isnull().sum()
```

```python
(df.isnull().sum()/len(df))*100
```

```python
## 1. Review Rating column has the missing value.
## 2. Fill null values with the median of their respective categories
```

```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
# Checking the null values again
df.isnull().sum()
```

```python
# Converting the columns name into snake casing

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

### Feature Enginerring

```python
# Creating a new column age_group to divide the customer in 4 categories for better understanding the patterns against each age group

# Define bins (age ranges)
bins = [0, 18, 35, 50, 65, 100]

# Define labels for those ranges
labels = ["Child", "Young Adult", "Adult", "Middle-Aged", "Senior"]

# Apply pd.cut
df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels, right=False)
```

```python
df[["age","age_group"]].head(10)
```

```python
# Creating a new column purchase_frequency_days to convert string data into numerical data for analysis
df["frequency_of_purchases"].unique()
```

```python

frequency_mapping = {
    'Fortnightly': 15,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```
